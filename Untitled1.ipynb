```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA

# --------------------------------------------------
# Step 1: Load the Iris Dataset
# --------------------------------------------------
data = load_iris()
X = data.data              # Feature matrix
y = data.target            # Class labels
target_names = data.target_names

print("Original Data Shape:", X.shape)

# --------------------------------------------------
# Step 2: Standardize the Data (Mean = 0, Std = 1)
# --------------------------------------------------
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# --------------------------------------------------
# Step 3: Normalize the Data (Range: 0 to 1)
# --------------------------------------------------
normalizer = MinMaxScaler()
X_normalized = normalizer.fit_transform(X_standardized)

# --------------------------------------------------
# Step 4: Apply PCA (Reduce to 2 Components)
# --------------------------------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)

# --------------------------------------------------
# Step 5: Display Explained Variance
# --------------------------------------------------
print("\nExplained Variance Ratio:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"Principal Component {i+1}: {var:.4f}")

print("Cumulative Explained Variance:",
      np.cumsum(pca.explained_variance_ratio_))

# --------------------------------------------------
# Step 6: 2D PCA Visualization
# --------------------------------------------------
plt.figure()
for i in np.unique(y):
    plt.scatter(
        X_pca[y == i, 0],
        X_pca[y == i, 1],
        label=target_names[i]
    )

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - 2D Visualization (Iris Dataset)")
plt.legend()
plt.show()

# --------------------------------------------------
# Step 7: Scree Plot (Cumulative Explained Variance)
# --------------------------------------------------
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_normalized)

plt.figure()
plt.plot(
    np.cumsum(pca_full.explained_variance_ratio_),
    marker='o'
)
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot (Cumulative Explained Variance)")
plt.grid(True)
plt.show()

# --------------------------------------------------
# Step 8: Original Data vs PCA Data Visualization
# --------------------------------------------------
plt.figure()

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Transformed Data")

plt.tight_layout()
plt.show()
```
